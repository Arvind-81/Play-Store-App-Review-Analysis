# Cleaning Dataset

```python
#There is one row has higher review
data_df[data_df['Rating']>5]
```

```python
data_df=data_df.drop(data_df.index[[10472]])
```

```python
data_df.describe()
```

```python
data_df['Reviews'].unique()
```

```python
data_df['Reviews']=pd.to_numeric(data_df['Reviews'],errors='coerce')
data_df['Reviews'].isnull().sum()
```

```python
# Null count in Each Column
for i in data_df.columns:
   print(i,'=',data_df[i].isna().sum(),'\n')
```

```python
 #Changing the type 'object' to 'float'
data_df=data_df.astype({"Reviews":float})
```

```python
data_df.info()
```

```python
#Checking Duplicate app
data_df.duplicated(['App']).value_counts()
```

```python
#Dropping Duplicate Value
data_df.drop_duplicates(['App'],inplace=True)
```

```python
data_df=data_df.drop(['Current Ver','Last Updated'],axis=1)
data_df
```

```python
#Drop NaN Values
data_df=data_df.dropna(subset=['Rating','Type','Content Rating','Category'])
data_df
```

```python
#Replace Price Column $ to''
data_df['Price']=data_df['Price'].str.replace('$','')
```

```python
data_df=data_df.astype({'Price':float})
```

```python
#Install column replace , and +
data_df['Installs']=data_df['Installs'].str.replace(',','')
data_df['Installs']=data_df['Installs'].str.replace('+','')
data_df=data_df.astype({'Installs':float})
```

```python
#Changing Size column object to float and in mb as well as drop some sign
data_df['Size']=data_df['Size'].apply(lambda x:str(x).replace('Varies with device', '20') if 'Varies with device' in str(x) else x)
data_df['Size']=data_df['Size'].apply(lambda x:str(x).replace('M', '') if 'M' in str(x) else x)
data_df['Size']=data_df['Size'].apply(lambda x:str(x).replace(',', '') if ',' in str(x) else x)
data_df['Size']=data_df['Size'].apply(lambda x:float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
data_df['Size']=data_df['Size'].apply(lambda x:float(x))

```

```python
#After Convergen
data_df['Size'].unique()
```

```python
#Final Dataset Look
data_df.head(5)
```

# Exploratory Analysis and Visualization 

```python
#Get the number of apps for each category
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
sns.countplot(x='Category',data=data_df)
plt.title('Number of Apps Per Category')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

```python
# 1. What is the percentage of each application type in Google play store?
x=data_df.groupby(['Category'],as_index=False)['App'].count()
print('Top Common App category in Google Play Store','\n',x.sort_values(by='App',ascending=False))
x.head()
```

```python
# 2. Which Category has the highest and the lower number of apps?
data_df['Category'].value_counts()
```

```python
#Reviewing the top 5 Categories
review=data_df.Category.value_counts().sort_values(ascending=False).head(5)
plt.figure(figsize=(15,5))
plt.xticks(rotation=75)
plt.style.use('bmh')
sns.barplot(x=review.index,y=review)
```

```python
# 3. Which Genre has highest number of install?
data_df['Genres'].value_counts()
```

```python
plt.figure(figsize=(10,5))
Genres=data_df['Genres'].value_counts().head(10)
Genres.plot.bar()
```

```python
# 4. What Percentage of App are Free or Paid in Playstore?
plt.figure(figsize=(7,7))
labels = data_df['Type'].value_counts(sort=True).index
sizes = data_df['Type'].value_counts(sort=True)
colors = ['yellow', 'purple']
explode = (0.2,0)
plt.pie(sizes , explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', startangle=0)
plt.title('FREE VS PAID APPS')
plt.show()
```

```python
# 5. Which Content rating columns are found more on playstore category wise?
x = data_df['Content Rating'].value_counts().index
y = data_df['Content Rating'].value_counts()
```

```python
plt.figure(figsize=(10,7))
plt.bar(x, y, width=0.4,  alpha=0.4);
plt.title('Content Rating',size=20);
```

```python
# 6. What are the top 10 expensive app in play store?
paid_10 =data_df['Type'] == 'Paid'
top_paid_apps =data_df[paid_10].sort_values(by='Price', ascending=False).head(11)
top_paid_apps
```

```python
# 7. What are the top 10 apps in terms of size?
max_size_apps =data_df.sort_values(by='Size', ascending=False).head(5)
max_size_apps
```

```python
# 8. How many apps are there whose rating is 4?
data_df[data_df['Rating'] == 4.5]
```

```python
# 9. Which apps are working on which android version?
data_df['Android Ver'].value_counts()
```

```python
# 10. Relationship between Rating & Reviews?
print('Apps having > 1 million reviews')
sns.jointplot(x='Reviews',y='Rating',data =data_df[data_df['Reviews'] > 100000],color='yellow', height=10)

plt.title('App reviews > 1M')
plt.style.use('dark_background')
plt.show()

print('Apps having < 1 million reviews')
sns.jointplot(x='Reviews',y='Rating',data =data_df[data_df['Reviews']<100000],color ='darkorange', height=10)

plt.title('App reviews < 1M')
plt.style.use('dark_background')
plt.show()
```

```python
# Show the distribution of app in term of their rating,size and type
plt.figure(figsize=(10,5))
sns.scatterplot(data_df['Size'],data_df['Rating'],hue=data_df['Type'],s=50)
```

```python
#Analyizing the numerical data columns and deriving corelation between them via heatmap?
sns.heatmap(data_df.corr(),annot=True,linewidths=1.0,fmt='.3f',square=True)
plt.title("Heatmap for numerical columns",size=30)
```

```python
# Sentiment Percentage
plt.figure(figsize=(10,5))
plt.title("Sentiment Percentage",fontsize=15)
plt.pie(user_df.Sentiment.value_counts(),labels=user_df.Sentiment.value_counts().index,autopct='%.2f%%',startangle=90,explode=(0,0.1,0.1))
plt.show()
```
